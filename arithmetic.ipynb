```python
def temperature_converter():

    temp = int(input('Enter your temperature value'))
    degree = int(input("choose the temperature unit \n 1. celcius \n 2. fahrenheit"))
    if degree == 1:
        print (temp,"degree farenheit to celsius is", int((temp-32)*5/9))
    elif degree == 2:
        print (temp, " degree celsius to farenheit to celsius is", int((9/5*temp)+ 32))
    elif degree == 3: 
        print (temp,"degree farenheit to Kelvin is", int((temp-32)*5/9 + 273.15))
    elif degree == 4:
         print (temp, " degree celsius to kelvin ", int((9/5*temp)+ 273.15))

    else:
        return "Invalid input, enter the correct input"
    
```
